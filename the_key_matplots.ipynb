# Key Matplots

In the Following Section we go through several kinds of plots that can be made with Matplotlib

````python
from matplotlib import pyplot as plt
````

````python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
````

## Bar Plots

### Side-by-Side Bar Plots

````python
plt.figure(figsize=(10,7), dpi=600)

#Paste the x_values code here
n = 1# This is our first dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar


#little trick to move the bars around for their postion on the x-axis
store1_x = [t*element + w*n for element
             in range(d)]

print(store1_x)

plt.bar(store1_x,sales1)

n = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store2_x = [t*element + w*n for element
             in range(d)]

print(store2_x)

plt.bar(store2_x, sales2)

plt.savefig("results/side_by_side_bar_plot.pdf")
plt.savefig("results/side_by_side_bar_plot.png", dpi=500)

plt.show()
````

### Stacked Barplots

````python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
````

For stacked barplots we specify the bottom paramater as done below to indicate that the one should be on top of the other one.

````python
plt.figure(figsize=(10,7), dpi=1200)
plt.bar(range(len(sales1)),
        sales1)

plt.bar(range(len(sales1)),
  sales2,
  bottom=sales1)

plt.legend(["Location 1", "Location 2"])

plt.savefig("results/stacked_bar_plot.pdf")

plt.show()
````

### Adding an Error Line

````python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]
error = [0.6, 0.9, 0.4, 0, 0.9, 0]
````

````python
# Plot the bar graph here
fig, ax = plt.subplots(dpi=1200, figsize=(10,7))

ax.bar(range(len(drinks)), ounces_of_milk, yerr=error, capsize=5) #Capsize is the hat/line on the ends of the error line

#You have to add both xticks and xticklabels to get them attached properly
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks)

#Trying to adjust fig size to subplot
#fig = plt.figure(figsize=(12,8))
#ax = fig.add_subplot()


plt.savefig("results/bar_with_std.png", dpi=1000)
plt.savefig("results/bar_with_std.pdf")
plt.show()
````

## Line plots with fill_between()

```
plt.fill_between(x_values, lower, upper, alpha_value)
```


````python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
````

````python
plt.figure(figsize=(10,7), dpi=300) #dpi can be set up to 2000
plt.plot(months, revenue)
ax = plt.subplot()

ax.set_xticks(months)
ax.set_xticklabels(month_names)

#to get 10% upper and lower bounds.
y_lower = [i-0.1*i for i in revenue]
y_upper = [i + 0.1 * i for i in revenue]

plt.fill_between(months, y_lower, y_upper, alpha=0.2)


plt.savefig("results/lineplot_err_line.png", dpi=300)
plt.savefig("results/lineplot_err_line.pdf")

#Always put show in the end when you want to .savefig to work. For some reason it doesn't work properly with
plt.show()
````

## Pie Chart (Leon's Favorite)

````python
from matplotlib import pyplot as plt


plt.figure(figsize=(8,8), dpi=500)
payment_method_names = ["Card Swipe", "Cash", "Apple Pay", "Other"]
payment_method_freqs = [270, 77, 32, 11]

#One way to display pie-chart details
plt.pie(payment_method_freqs,
        labels=payment_method_names,
        autopct='%0.1f%%')

#Another way to display pie-charts details / names
plt.legend(payment_method_names)
plt.axis('equal')


plt.savefig("results/pie_chart.png", dpi=300)
plt.savefig("results/pie_chart.pdf")
plt.show()

````
