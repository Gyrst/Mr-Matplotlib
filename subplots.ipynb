# Subplots using line graphs

```python
from matplotlib import pyplot as plt
```

### Data

```python

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]

```

The figure is the overall page that we are creating. This may include one or more subplots

+ You can create as many subplots with the main plots as desired specificying plt.subplot(rows,cols,the_plot_you_want_to_edit), where the count of plots start at 1 and not index 0.

```python
#subplot example
plt.figure(figsize=(10,10))
rows, cols, position = 2,2,1
plt.subplot(rows,cols,position)

#Added the second
plt.subplot(rows,cols,2)

#Added fourth position - the rightmost lower corner in this case
plt.subplot(rows,cols,4)
```

```python
# create your figure here 
plt.figure(figsize=(12,8))

#Left Subplot
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='o')
ax1.set_xticklabels(months)
ax1.set_xlabel("Months")
ax1.set_ylabel("Visits")
ax1.set_title("Presentation of Visits Over Time")


#Right Subplot
ax2 = plt.subplot(1,2,2)

plt.plot(x_values, key_limes_per_month)
plt.plot(x_values, persian_limes_per_month)
plt.plot(x_values,blood_limes_per_month)
ax2.legend(["Key lime", "Persion lime", "Blood lime"])
ax2.set_xticklabels(months)
ax2.set_xlabel("Months")
ax2.set_ylabel("Sales")
ax2.set_title("Breakdown of Lime Sales Over Time")

#adjusting margins etc. 
plt.subplots_adjust(bottom=0.3, right=1, top=1)

plt.savefig("results/subplot_limes.png", dpi=300)
```
